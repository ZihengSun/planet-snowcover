```python
import matplotlib 
import matplotlib.pyplot as plt 
import cartopy.crs as ccrs
import cartopy.feature as feature
import geopandas as gpd

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
states_feature = feature.NaturalEarthFeature(category='cultural',
                                             name='admin_1_states_provinces_lakes_shp', 
                                             scale='10m',
                                             facecolor='none', 
                                            edgecolor='black')


stations = gpd.read_file('sites.shp')
stations.crs = '4326'
```

```python
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection = ccrs.Mercator().GOOGLE)
#ax.set_extent([-127, -116, 45, 50], crs=ccrs.PlateCarree())
ax.add_feature(states_feature)
ax.gridlines(draw_labels=True)
stations.plot(column='elev', cmap='BuPu', scheme='quantiles', legend=True, ax = ax, marker = 'o', markersize=10)

plt.show()
```
