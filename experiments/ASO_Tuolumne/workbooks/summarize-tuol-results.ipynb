```python
resultsData = pd.concat([pd.DataFrame(json.load(open(f))).T for f in results])
```

```python
resultsData.T.reset_index()
```

```python
tilenos = [int(open("/" + os.path.join(*result.split("/")[:-2] + ["tileno.txt"])).read().strip()) for result in results]
```

```python
predictions = resultsData[resultsData.index.str.contains("merged")]
sentinel = resultsData[resultsData.index.str.contains("NDSI")]
landsat = resultsData[resultsData.index.str.contains("SNOW")]
```

```python
predictions['tileno'] = tilenos
sentinel['tileno'] = tilenos
landsat['tileno'] = tilenos
```

```python
predictions = predictions.reset_index(drop='True')
predictions['SCA'] = 'CubeSat'
```

```python
sentinel = sentinel.reset_index(drop="True")
sentinel['SCA'] = "Sentinel-2"
```

```python
landsat = landsat.reset_index(drop='True')
landsat['SCA'] = "Landsat 8 fSCA"
```

```python
merged = pd.concat([predictions, sentinel, landsat])
```

```python
merged
```

```python
# melted2 = pd.melt(merged[merged.tileno > MIN_TILENO].drop('tileno', axis=1).reset_index(), id_vars=['index', 'SCA'])
melted2 = pd.melt(merged.drop('tileno', axis=1).reset_index(), id_vars=['index', 'SCA'])
melted2 = melted2[melted2.value > 0]
```

```python
box_pairs = [
    (("Precision", "CubeSat"),("Precision", "Sentinel-2")),
    (("Precision", "CubeSat"),("Precision", "Landsat 8 fSCA")),
    
    (("Recall", "CubeSat"),("Recall", "Sentinel-2")),
    (("Recall", "CubeSat"),("Recall", "Landsat 8 fSCA")),

    (("F-Score", "CubeSat"),("F-Score", "Sentinel-2")),
    (("F-Score", "CubeSat"),("F-Score", "Landsat 8 fSCA")),

    (("Balanced Accuracy", "CubeSat"),("Balanced Accuracy", "Sentinel-2")),
    (("Balanced Accuracy", "CubeSat"),("Balanced Accuracy", "Landsat 8 fSCA")),

    (("Kappa", "CubeSat"),("Kappa", "Sentinel-2")),
    (("Kappa", "CubeSat"),("Kappa", "Landsat 8 fSCA")),    
]
```

```python

from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
sns.set(style='ticks', font_scale=1.5)

fig, ax = plt.subplots(figsize=(11.5,5), dpi=300)
bp = sns.boxplot(x='variable', hue='SCA', y = 'value', data=melted2, ax= ax)
plt.ylim([0, 1])
plt.xlabel("")
plt.ylabel("")
# ax.yaxis.set_minor_locator(AutoMinorLocator(2))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
sns.despine()
plt.tight_layout()
plt.savefig("tuol-sentinel-ml-performance.png", dpi=300, bbox_inches='tight')
test_results = add_stat_annotation(ax, data=melted2, x='variable', y='value', hue='SCA',
                                   box_pairs=box_pairs,
                                   test='t-test_ind', text_format='star',
                                   loc='outside', verbose=2)
plt.savefig("tuol-sentinel-ml-performance_sig.png", dpi=300, bbox_inches='tight')
```
